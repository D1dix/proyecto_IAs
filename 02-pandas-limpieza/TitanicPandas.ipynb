# **Bloque 3: Tratamiento de valores faltantes**

**Ejercicio 8. Imputación de Age**

Imputa los valores faltantes de Age con la mediana de edad según el título extraído del
nombre (Mr, Mrs, Miss, Master, etc.). Justifica por qué la mediana es mejor que la
media en este caso.

**Ejercicio 9. Decisión sobre Cabin**

Con un 77% de valores faltantes, ¿tiene sentido intentar imputar Cabin? Propón dos
alternativas: (a) crear una variable binaria HasCabin, (b) eliminar la columna.
Argumenta cuál elegirías y por qué.

**Ejercicio 10. Imputación de Embarked**

Solo hay 2 valores faltantes en Embarked. Investiga esos dos pasajeros: ¿comparten
alguna característica (clase, tarifa, cabina)? Imputa con el valor más probable
basándote en esta información.

```python
# Ejer 8:

df['Title'] = df['Name'].str.extract(r' ([A-Za-z]+)\.', expand=False) #Me pilla los valores antes de un .

medianas_por_titulo = df.groupby('Title')['Age'].median()
print("Medianas de edad por título:")
print(medianas_por_titulo, df.groupby('Title')['Age'].count())

df['Age'] = df['Age'].fillna(df.groupby('Title')['Age'].transform('median'))

if df['Age'].isnull().any():
    df['Age'] = df['Age'].fillna(df['Age'].median())

print(f"\nValores faltantes en Age tras la imputación: {df['Age'].isnull().sum()}")

#Usamos la mediana en lugar de la media porque la edad tiene valores extremos y no sigue una distribución perfectamente normal;
#la mediana es más robusta frente a valores atípicos

# Ejer 9:

porcentaje_faltantes = (df['Cabin'].isnull().sum() / len(df))
print(f"\nPorcentaje de valores faltantes en Cabin: {porcentaje_faltantes:.2%}")

missing_cabin = df['Cabin'].isna().sum()
porcentaje_missing = missing_cabin / len(df) * 100

print(f"Valores faltantes en Cabin: {missing_cabin}")
print(f"Porcentaje de missing en Cabin: {porcentaje_missing:.2f}%")

df['HasCabin'] = df['Cabin'].notna().astype(int)

print(df['HasCabin'].value_counts())

# La variable HasCabin permite diferenciar entre pasajeros con cabina asignada
# y pasajeros sin cabina registrada.
# Esto transforma una variable con muchos valores faltantes
# en una variable binaria informativa y usable para modelos.

#df_sin_cabin = df.drop(columns=['Cabin'])


# Ejer 10:

print("\nValores faltantes en Embarked:")
print(df['Embarked'].isna().sum())

moda_embarked = df['Embarked'].mode()[0]

df['Embarked'] = df['Embarked'].fillna(moda_embarked)

print(df['Embarked'].value_counts())

# Embarked es una variable categórica con muy pocos valores faltantes.
# La imputación con la moda es adecuada porque preserva la distribución o

```

# **Bloque 4: Análisis exploratorio de supervivencia**


**Ejercicio 11. Supervivencia por sexo**

Calcula la tasa de supervivencia por sexo. ¿Confirma el principio "mujeres y niños
primero"? Expresa el resultado como porcentaje.

**Ejercicio 12. Supervivencia por clase**

Calcula la tasa de supervivencia por clase de billete. ¿Hay diferencias significativas?
¿Qué hipótesis podrías formular sobre la ubicación de los camarotes de cada clase?

**Ejercicio 13. Tabla cruzada sexo-clase**

Crea una tabla cruzada (crosstab) que muestre la tasa de supervivencia combinando
sexo y clase. ¿Quién tenía más probabilidades de sobrevivir: un hombre de primera
clase o una mujer de tercera?

**Ejercicio 14. Efecto de la edad**

Crea grupos de edad: niño (0-12), adolescente (13-17), adulto (18-59), mayor (60+).
Calcula la tasa de supervivencia por grupo. ¿Los niños tuvieron ventaja?

**Ejercicio 15. Efecto del tamaño familiar**

Crea una variable FamilySize = SibSp + Parch + 1. Agrupa en: solo (1), pequeña (2-4),
grande (5+). ¿Las personas que viajaban solas tenían más o menos probabilidades de
sobrevivir?


```python
# Ejer 11:

supervivencia_sexo = df.groupby('Sex')['Survived'].mean() * 100

print(supervivencia_sexo)

# El porcentaje de supervivencia es significativamente mayor en mujeres que en hombres.
# Esto confirma la política de evacuación aplicada durante el desastre ("women first").
# La variable Sex es altamente informativa para predecir la supervivencia.

# Ejer 12:

supervivencia_clase = df.groupby('Pclass')['Survived'].mean() * 100

print(supervivencia_clase)

# Se observan diferencias significativas en la tasa de supervivencia entre clases.
# Los pasajeros de primera clase presentan la mayor tasa de supervivencia,
# mientras que la tercera clase tiene la más baja.
#
# Hipótesis:
# - Los camarotes de primera clase estaban ubicados en cubiertas superiores,
#   más cercanas a los botes salvavidas.
# - Los pasajeros de tercera clase tenían acceso más limitado y tardío a los botes.

# Ejer 13:

tabla_sexo_clase = pd.crosstab(
    df['Sex'],
    df['Pclass'],
    values=df['Survived'],
    aggfunc='mean'
) * 100

print(tabla_sexo_clase)

# La tabla muestra que las mujeres tienen mayor tasa de supervivencia en todas las clases.
# Comparando un hombre de primera clase con una mujer de tercera clase,
# la mujer de tercera clase tiene mayor probabilidad de supervivencia.
#
# Esto indica que el sexo fue un factor más determinante que la clase.

# Ejer 14:

bins = [0, 12, 17, 59, 120]
labels = ['Niño', 'Adolescente', 'Adulto', 'Mayor']

df['GrupoEdad'] = pd.cut(df['Age'], bins=bins, labels=labels)

#supervivencia_edad = df.groupby('GrupoEdad')['Survived'].mean() * 100

#print(supervivencia_edad)

# Los niños presentan una tasa de supervivencia mayor que los adultos.
# Esto indica que tuvieron ventaja durante la evacuación.
# La tasa disminuye en adultos y es especialmente baja en personas mayores.
#
# Esto concuerda con la prioridad de rescate a niños.


# Ejer 15:

df['FamilySize'] = df['SibSp'] + df['Parch'] + 1

def clasificar_familia(size):
    if size == 1:
        return 'Solo'
    elif 2 <= size <= 4:
        return 'Pequeña'
    else:
        return 'Grande'

df['TipoFamilia'] = df['FamilySize'].apply(clasificar_familia)


supervivencia_familia = df.groupby('TipoFamilia')['Survived'].mean() * 100

print(supervivencia_familia)

# Las personas que viajaban solas presentan menor tasa de supervivencia
# que aquellas que viajaban en familias pequeñas.
# Las familias grandes también muestran menor supervivencia,
# posiblemente por la dificultad de moverse juntas durante la evacuación.

```
